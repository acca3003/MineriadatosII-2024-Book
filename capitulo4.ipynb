Existen diferentes modelamientos de consecuentes, los m√°s destacados son:

$$ y = \frac{{\sum_{i=1}^{n} \mu_{A_i}(x_i) \cdot \text{V}_i}}{{\sum_{i=1}^{n} \mu_{A_i}(x_i)}} $$

donde $y$ es el valor de salida, $\mu_A(x)$ es la funci√≥n de pertenencia difusa del antecedente $A_i$ evaluada en el valor de entrada, $V_i$ es el valor asociado al consecuente para el antecedente $A_i$ y *n* es el n√∫mero de antecedentes en la regla

Esta f√≥rmula calcula el valor de salida *ùë¶* como una combinaci√≥n ponderada de los valores asociados $V_i$ para cada antecedente, donde los pesos est√°n determinados por las funciones de pertenencia de los antecedentes. La suma de las funciones de pertenencia de los antecedentes normaliza el resultado, asegurando que est√© en el rango adecuado.

> Estas f√≥rmulas proporcionan una manera matem√°tica de expresar los antecedentes y consecuentes en un sistema de l√≥gica difusa, lo que permite modelar y controlar sistemas basados en reglas ling√º√≠sticas y condiciones difusas.

La salida de un sistema difuso es, por tanto, un conjunto difuso. En cambio, los sistemas de control se relacionan con el mundo externo a trav√©s de valores exactos lo que implica, a la hora de hacer inferencia, es necesario realizar un proceso de **defuzzificaci√≥n**. As√≠, los principales m√©todos de defuzzificaci√≥n son:

-   `Centroide`: calcula el centro de gravedad del conjunto difuso ponderado. Consiste en encontrar el punto en el eje de salida donde el √°rea bajo la curva del conjunto difuso es dividida en dos √°reas iguales. Matem√°ticamente, el valor de salida *y* se calcula como

::: {style="text-align:center;"}
$$ y = \frac{{\sum_{i=1}^{n} u_i \cdot x_i}}{{\sum_{i=1}^{n} u_i}} $$
:::

-   `M√°ximo Valor de Membres√≠a`: selecciona el valor de salida con el grado de membres√≠a m√°s alto. Es decir, el valor de salida es el punto en el conjunto difuso donde la membres√≠a es m√°xima. As√≠, $y = max(u(x)$

-   `Media de los M√°ximos`: se calcula la media de todos los puntos en el conjunto difuso donde el grado de membres√≠a es m√°ximo. Es √∫til cuando hay m√∫ltiples picos en el conjunto difuso

-   `Primera M√°xima y √öltima M√°xima`: se selecciona el primer y √∫ltimo punto de m√°ximo grado de membres√≠a, respectivamente, como valores de salida

-   `Bisectriz`: este m√©todo divide el conjunto difuso en dos √°reas de igual tama√±o y toma el punto medio entre los dos puntos de intersecci√≥n de la bisectriz con el conjunto difuso

## Ejemplos pr√°cticos

Para ilustrar un ejemplo de control de sistemas difuso, consideremos un sistema difuso simple para controlar la potencia de un calefactor basado en la temperatura ambiente. Supongamos que queremos mantener la temperatura en una habitaci√≥n alrededor de un valor de referencia.

**Definici√≥n de las Variables Difusas** - `Temperatura`: esta variable representa la temperatura medida en grados Celsius. Creamos una variable de entrada temperatura que abarca valores desde 0 hasta 100 grados Celsius. - `Potencia`: esta variable representa el nivel de potencia de la calefacci√≥n. Creamos una variable de salida potencia que tambi√©n abarca valores desde 0 hasta 100.

**Funciones de Membres√≠a** - `Temperatura`: se definen tres funciones de membres√≠a para la temperatura: "fr√≠a", "templada" y "caliente". - `Potencia`: de igual forma, se definen tres funciones de membres√≠a para la potencia de la calefacci√≥n: "baja", "media" y "alta".

**Reglas difusas** Se define el sistema de reglas difusas: - Si la temperatura es fr√≠a, entonces la potencia debe ser alta - Si la temperatura es alta, entonces la potencia debe ser baja - Si la temperatura es templada, entonces la potencia debe ser media

**Sistema de Control** Se crea un conjunto de reglas difusas a partir de cada una de las reglas definidas previsamente. Este conjunto de reglas representa el conocimiento difuso que guiar√° el comportamiento del sistema de control.

**Controlador Difuso** Se crea un controlador difuso que toma como argumento el sistema de control difuso creado anteriormente. Este controlador simula el comportamiento del sistema de control difuso y nos permite introducir valores de entrada (temperatura) para obtener el valor de salida (potencia) correspondiente.

**Simulaci√≥n** Se simula un cambio en la temperatura estableciendo un valor para la variable de entrada temperatura en el controlador difuso. Luego, se calcula el valor de salida (nivel de potencia) utilizando el m√©todo compute() del controlador.

Finalmente, el c√≥digo python:


````python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Definici√≥n de las variables de entrada y salida
temperatura = ctrl.Antecedent(np.arange(0, 101, 1), 'temperatura')
potencia = ctrl.Consequent(np.arange(0, 101, 1), 'potencia')

# Definici√≥n de las funciones de membres√≠a para la temperatura
temperatura['fr√≠a'] = fuzz.trimf(temperatura.universe, [0, 0, 50])
temperatura['templada'] = fuzz.trimf(temperatura.universe, [0, 50, 100])
temperatura['caliente'] = fuzz.trimf(temperatura.universe, [50, 100, 100])

# Definici√≥n de las funciones de membres√≠a para la potencia
potencia['baja'] = fuzz.trimf(potencia.universe, [0, 0, 50])
potencia['media'] = fuzz.trimf(potencia.universe, [0, 50, 100])
potencia['alta'] = fuzz.trimf(potencia.universe, [50, 100, 100])

# Reglas difusas
regla1 = ctrl.Rule(temperatura['fr√≠a'], potencia['alta'])
regla2 = ctrl.Rule(temperatura['templada'], potencia['media'])
regla3 = ctrl.Rule(temperatura['caliente'], potencia['baja'])

# Sistema de control difuso
sistema_control = ctrl.ControlSystem([regla1, regla2, regla3])
controlador = ctrl.ControlSystemSimulation(sistema_control)

# Simulaci√≥n de un cambio en la temperatura
controlador.input['temperatura'] = 25  # Temperatura medida en grados Celsius

# Computar el resultado
controlador.compute()

# Visualizaci√≥n del resultado
print("Nivel de potencia de la calefacci√≥n:", controlador.output['potencia'])
````

A continuaci√≥n, se representan e interpretan la funciones de membres√≠a antecedente y consecuente del sistema de control:


````python
# Detalle de las funciones de membres√≠a (antecedente y consecuente)

temperatura.view(sim=controlador) 
````

````python
potencia.view(sim=controlador)
````

## Fuzzy C-Means

Los algoritmos de clustering difuso, como el Fuzzy C-Means (FCM), son esenciales en la agrupaci√≥n de datos donde la pertenencia de un punto a un cl√∫ster no es binaria, sino que se modela como un grado de pertenencia difuso. FCM asigna a cada punto un grado de pertenencia a todos los cl√∫steres, lo que permite una representaci√≥n m√°s flexible de la estructura de los datos. En lugar de asignar cada punto a un √∫nico cl√∫ster, FCM asigna grados de pertenencia que indican la probabilidad de que un punto pertenezca a cada cl√∫ster. Esto es especialmente √∫til cuando los datos pueden pertenecer a m√∫ltiples cl√∫steres simult√°neamente, o cuando la frontera entre cl√∫steres no es clara. La importancia de los algoritmos de clustering difuso radica en su capacidad para manejar la incertidumbre en los datos y proporcionar una agrupaci√≥n m√°s completa y flexible que los m√©todos tradicionales de clustering.

> A diferencia del algoritmo C-Means cl√°sico, que trabaja con una partici√≥n dura, FCM realiza una partici√≥n suave del conjunto de datos. En tal partici√≥n los datos pertenecen en alg√∫n grado a todos los clusters.

El siguiente c√≥digo muestra c√≥mo hacer un C-Means difuso:


````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from sklearn import datasets

# Carga de datos de Iris desde scikit-learn
iris = datasets.load_iris()
X = iris.data  # Caracter√≠sticas
y = iris.target  # Etiquetas

# Gr√°fico de dispersi√≥n para visualizar los clusters originales seg√∫n las dos primeras caracter√≠sticas (s√©palo)
g = sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette="Set2")
plt.title('S√©palo: longitud vs tama√±o'), plt.xlabel('Longitud'), plt.ylabel('Ancho')
plt.show()

# Algoritmo Fuzzy C-Means
# Es necesario que los datos est√©n en un array 2D, por lo que se utiliza la funci√≥n "reshape"
X2 = np.reshape(X.T, (X.shape[1], X.shape[0]))  # Cambio de forma de los datos

# Aplicaci√≥n del modelo Fuzzy C-Means
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(X2, c=3, m=2, error=0.005, maxiter=1000, init=None, seed=111)
````

Entre los par√°metros que devuelve la funci√≥n `fuzz.cluster.cmeans` se pueden citar: - `cntr` es el centro de los clusters - `u` es el grado de membres√≠a - `u0` es la matriz inicial de membres√≠a - `d` es la matriz de distancias euclidiana


````python
# Centro de los clusters
cluster_centers_df = pd.DataFrame(cntr, columns=['Longitud S√©palo', 'Ancho S√©palo', 'Longitud P√©talo', 'Ancho P√©talo'],
                                  index=['Cluster 1', 'Cluster 2', 'Cluster 3'])
cluster_centers_df
````

````python
# Grados de membres√≠a (probabilidad de pertenencia a cada cluster)
membership_df = pd.DataFrame(u.T, columns=['Cluster 1', 'Cluster 2', 'Cluster 3'])
membership_df.head()
````

````python
# Gr√°fico de dispersi√≥n para visualizar los clusters obtenidos
g = sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=np.argmax(u.T, axis=1), palette="Set2")  # Asignaci√≥n de colores seg√∫n cluster
plt.title('S√©palo: longitud vs tama√±o'), plt.xlabel('Longitud'), plt.ylabel('Ancho')
plt.show()
````

## Fuzzy Matching

El **Fuzzy Matching** es una t√©cnica utilizada para comparar cadenas de texto y determinar su grado de similitud, incluso cuando hay diferencias ortogr√°ficas, tipogr√°ficas u otras variaciones entre ellas. Esta t√©cnica identifica la probabilidad de que dos registros sean realmente coincidentes en funci√≥n de si coinciden o no en diferentes identificadores. Los identificadores elegidos y el peso asignado constituyen la base de la *Concordancia Difusa*. As√≠, cuando los par√°metros son demasiado amplios, se encontrar√°n m√°s coincidencias, pero tambi√©n aumentar√°n las posibilidades de "falsos positivos".

Existen diversas m√©tricas a utilizar en Fuzzy Matching siendo la m√°s destacada la `distancia de Levenshtein` pero existen otras como se indican a continuaci√≥n:

-   `Distancia de Levenshtein`: mide el n√∫mero m√≠nimo de operaciones necesarias para convertir una cadena en otra. Las operaciones incluyen inserci√≥n, eliminaci√≥n o sustituci√≥n de caracteres
-   `Distancia Damerau-Levenshtein`: similar a la distancia de Levenshtein, pero tambi√©n considera la transposici√≥n de caracteres adyacentes como una operaci√≥n v√°lida
-   `Distancia Jaro-Winkler`: dise√±ada para comparar cadenas de texto cortas, tiene en cuenta la frecuencia de los caracteres y la posici√≥n de las coincidencias comunes
-   `Distancia del teclado`: eval√∫a la similitud entre dos cadenas bas√°ndose en la proximidad de las teclas en un teclado est√°ndar
-   `Distancia Kullback-Leibler`: utilizada en comparaciones de cadenas de texto que representan distribuciones de probabilidad

> La **concordancia difusa** es crucial para identificar similitudes entre registros, permitiendo la gesti√≥n eficiente de datos duplicados. La distancia de Levenshtein es fundamental en este proceso, al calcular la diferencia entre dos cadenas de caracteres, siendo especialmente √∫til para detectar errores ortogr√°ficos y variaciones en la escritura. M√°s sobre Distancia de Levenshtein en el siguiente enlace: <https://es.wikipedia.org/wiki/Distancia_de_Levenshtein>

### Ejemplo 1


````python
from fuzzywuzzy import fuzz

# Funci√≥n para comparar nombres completos
def comparar_nombres(nombre1, nombre2):
    # Usamos el ratio completo para comparar nombres
    ratio = fuzz.ratio(nombre1.lower(), nombre2.lower())
    return ratio

# Ejemplo de comparaci√≥n de nombres
nombre1 = "Juan P√©rez"
nombre2 = "Juan P√©rez G√≥mez"
nombre3 = "Pedro L√≥pez"
nombre4 = "Mar√≠a P√©rez G√≥mez"

print("Comparaci√≥n de nombres completos:")
print(f"{nombre1} vs {nombre2}: {comparar_nombres(nombre1, nombre2)}")
print(f"{nombre1} vs {nombre3}: {comparar_nombres(nombre1, nombre3)}")
print(f"{nombre2} vs {nombre4}: {comparar_nombres(nombre2, nombre4)}")
````

### Ejemplo 2


````python
# Funci√≥n para comparar direcciones de vivienda
def comparar_direcciones(direccion1, direccion2):
    # Usamos el ratio parcial para comparar direcciones
    ratio = fuzz.partialratio(direccion1.lower(), direccion2.lower())
    return ratio

# Ejemplo de comparaci√≥n de direcciones
direccion1 = "123 Calle Principal, Madrid"
direccion2 = "245 Calle Principal, Madrid"
direccion3 = "456 Calle Secundaria, √Åvila"
direccion4 = "123 Calle Principal, √Åvila"

print("\nComparaci√≥n de direcciones de vivienda:")
print(f"{direccion1} vs {direccion2}: {comparar_direcciones(direccion1, direccion2)}")
print(f"{direccion1} vs {direccion3}: {comparar_direcciones(direccion1, direccion3)}")
print(f"{direccion2} vs {direccion4}: {comparar_direcciones(direccion2, direccion4)}")
print(f"{direccion3} vs {direccion4}: {comparar_direcciones(direccion3, direccion4)}")
````

En la biblioteca **fuzzywuzzy**, las funciones `ratio()` y `partial_ratio()` se utilizan para calcular la similitud entre dos cadenas de texto. La diferencia principal radica en c√≥mo se realiza la comparaci√≥n y qu√© tan flexible es cada m√©todo en t√©rminos de coincidencia.

-   `ratio()`: esta funci√≥n calcula la similitud entre dos cadenas compar√°ndolas en su totalidad. Compara las cadenas car√°cter por car√°cter y devuelve un valor que representa la similitud entre las dos cadenas en t√©rminos de coincidencia exacta. Es √∫til cuando se busca una coincidencia exacta entre las cadenas y se prefiere una comparaci√≥n estricta.
-   `partial_ratio()`: esta funci√≥n calcula la similitud entre dos cadenas tomando en cuenta solo una parte de las mismas. Se utiliza para buscar subcadenas que coincidan parcialmente entre las cadenas de texto. Esto significa que puede ser m√°s √∫til cuando se enfrenta a casos donde las cadenas pueden tener variaciones menores o cuando se desea encontrar coincidencias incluso si las cadenas difieren en longitud o tienen diferencias menores.

Entonces, para decidir cu√°l usar, considera la naturaleza de los datos y el nivel de flexibilidad para realizar la comparaci√≥n. Suele usarse `ratio()` cuando se necesita una comparaci√≥n estricta y quieras asegurarte de que las cadenas coincidan exactamente; mientras que `partial_ratio()` se emplea cuando se buscan coincidencias parciales o flexibles entre las cadenas (p.e: cuando solo importan importa una parte espec√≠fica de las mismas).
